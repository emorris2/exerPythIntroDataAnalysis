```python
from matplotlib import pyplot as plt
```

```python
x = [1, 2, 3]
y = [1, 4, 9]
plt.plot(x, y)
plt.show()
```

```python
x = [1, 2, 3]
y = [1, 4, 9]
plt.plot(x, y)
plt.title("test plot")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
x = [1, 2, 3]
y = [1, 4, 9]
z = [10, 5, 0]
plt.plot(x, y)
plt.plot(x, z)
plt.title("test plot")
plt.xlabel("x")
plt.ylabel("y and z")
plt.show()
```

```python
x = [1, 2, 3]
y = [1, 4, 9]
z = [10, 5, 0]
plt.plot(x, y)
plt.plot(x, z)
plt.title("test plot")
plt.xlabel("x")
plt.ylabel("y and z")
plt.legend(["this is y", "this is z"])
plt.show()
```

```python
data = pd.read_csv('countries.csv')
```

```python
data
```

```python
# Compare the population growth in the US and China
```

```python
data.country == 'United States'
```

```python
data[data.country == 'United States']
```

```python
us = data [data.country == 'United States']
```

```python
china = data[data.country == 'China']
```

```python
china
```

```python
us
```

```python
plt.plot(us.year, us.population)
plt.show()
```

```python
plt.plot(us.year, us.population / 10**6)
plt.show()
```

```python
plt.plot(us.year, us.population / 10**6)
plt.plot(china.year, china.population / 10**6)
plt.show()
```

```python
plt.plot(us.year, us.population / 10**6)
plt.plot(china.year, china.population / 10**6)
plt.legend(['United States', 'China'])
plt.xlabel('year')
plt.ylabel('population')
plt.show()
```
